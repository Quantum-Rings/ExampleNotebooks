## 2. Build and Execute a Circuit

```python
import pennylane as qml

dev = qml.device(
    "openquantum.device",
    wires=2,
    shots=100,
    backend="iqm:emerald",
    client_id=CLIENT_ID,
    client_secret=CLIENT_SECRET
)

@qml.qnode(dev)
def bell_state():
    qml.Hadamard(wires=0)
    qml.CNOT(wires=[0, 1])
    return qml.counts()

result = bell_state()
print(result)
# {'00': 507, '11': 517}
```
