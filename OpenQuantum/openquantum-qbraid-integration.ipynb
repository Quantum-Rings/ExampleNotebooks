## 3. Build and Execute a Circuit

The qBraid runtime supports 10+ languages, so you can write circuits using your preferred framework. This example uses Qiskit.

```python
from qiskit import QuantumCircuit

circuit = QuantumCircuit(2, 2)
circuit.h(0)
circuit.cx(0, 1)
circuit.measure([0, 1], [0, 1])

job = device.run(circuit, shots=10)
result = job.result()
print(result.data.get_counts())
# {'00': 507, '11': 517}
```
