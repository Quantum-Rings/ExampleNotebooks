# Qiskit on Open Quantum

The simplest possible Qiskit example on Open Quantum: build a Bell state
and run it on real quantum hardware. Every line comes straight from the
[Qiskit plugin documentation](https://docs.openquantum.com/qiskit-plugin/getting-started/).

## Connect

Paste in an SDK key from your
[Open Quantum account](https://www.openquantum.com/settings/keys).


```python
from openquantum_sdk.auth import ClientCredentials
from openquantum_sdk_qiskit import OpenQuantumService

service = OpenQuantumService(
    creds=ClientCredentials(
        client_id="s_your_client_id",
        client_secret="your_client_secret",
    ),
)
```

## Run a Bell state on real hardware


```python
from qiskit.circuit import QuantumCircuit

# Get a backend
backend = service.return_backend("ionq:forte-1")

# Build a Bell state circuit
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

# Run on Open Quantum
job = backend.run(qc, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
```

A Bell state entangles the two qubits, so the counts land almost
entirely on `00` and `11`, with a few other results from hardware noise.
