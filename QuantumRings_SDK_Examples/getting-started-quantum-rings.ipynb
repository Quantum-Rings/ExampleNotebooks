Initialize the Quantum Rings backend. Use **scarlet_quantum_rings** for CPU mode, **amber_quantum_rings** for GPU mode, and **serin_quantum_rings** for hybrid mode.

```python
provider = QuantumRingsProvider(token=ACCOUNT_TOKEN, name=ACCOUNT_NAME)
backend = provider.get_backend("scarlet_quantum_rings")
```

## 3. Build and Execute a Circuit
Build a circuit using the native Quantum Rings library, or import the [Quantum Rings Qiskit Plugin](https://portal.quantumrings.com/doc/integrations/toolkit_primitives.html) and build your circuits in qiskit.

```python
from QuantumRingsLib import QuantumCircuit, job_monitor
import qiskit

# Bell Circuit
qc = QuantumCircuit(2,2) # 2 qbits, 2 classical bits
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

# Execute the circuit on the Quantum Rings backend
job = backend.run(qc, shots=100)
job_monitor(job)
result = job.result()
print(result.get_counts())

# Plot Results
from qiskit.visualization import plot_histogram
plot_histogram(result.get_counts(), figsize=(6,4), title="Job Results")
```
